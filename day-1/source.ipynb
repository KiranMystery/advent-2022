```python
with open('calorie.txt') as f:
    lines = f.readlines()
```

```python
i = 0
new_list=[]
first_try=True
for element in lines:
    print(i, element)
    if element == '\n':
        i+=1
        first_try = True
        continue
    else:
        if first_try == False:
            new_list[i]+=int(element.replace('\n',''))
        else:
            new_list.append(int(element.replace('\n','')))
            first_try = False
```
