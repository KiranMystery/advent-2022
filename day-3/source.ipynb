```python
total = 0
dupe = False
for x in X:
    compartment_1 = slice(0, len(x)//2)
    compartment_2 = slice(len(x)//2, len(x))
    print(x[compartment_1], x[compartment_2])
    
    for c in x[compartment_1]:
        print(c)
        if c in x[compartment_2]:
            print(c)
            dupe_char = c
            break        
    if dupe_char.islower():
        priority = ord(dupe_char) - 96
    else:
        priority = ord(dupe_char) - 38

    total += priority
```

```python
def pairwise(iterable):
    a = iter(iterable)
    return zip(a, a, a)
    
```

```python
total = 0
    
for elf_1, elf_2, elf_3 in pairwise(X):
    for c in elf_1:
        if c in elf_2 and c in elf_3:
                print('elf 3:',c)
                dupe_char = c
                break
                
    if dupe_char.islower():
        priority = ord(dupe_char) - 96
    else:
        priority = ord(dupe_char) - 38

    total += priority
```
