```python
input = [[int(l) for l in l.strip()] for l in open('file.in')]

X=np.array(input)

l = len(X)
visible_trees=4*(l-1)
trees_processed = visible_trees
print("outer trees:", visible_trees)
for row_i, row_x in enumerate(X):
    # ignore first and last row
    if row_i==0 or row_i==len(X)-1:
        continue
    else:
        # ignore first and last column
        for i, x in enumerate(row_x):
            if i==0 or i==len(row_x)-1:
                continue
            else:
                trees_processed += 1

                left_max=row_x[:i].max()
                right_max=row_x[i+1:].max()
                top_max=X[:, row_i][:i].max()
                bottom_max=X[:, row_i][i+1:].max()
                
                
                if x>left_max or \
                    x>right_max or \
                        x>top_max or \
                            x>bottom_max:
                    print(f"({row_i}, {i}) {x} is greater than; left {left_max}, right {right_max}, top {top_max}, bottom {bottom_max}:",True)

                    visible_trees += 1
                else:
                    print(f"({row_i}, {i}) {x} is greater than; left {left_max}, right {right_max}, top {top_max}, bottom {bottom_max}:",False)


print("ending trees:", visible_trees)
print("trees processed:", trees_processed)
```
